QUESTION 1:
What are Large Language Models (LLMs) and how do they function?

ANSWER:
Large Language Models (LLMs) are deep learning models trained on massive amounts of text data to understand and generate human language.

They function using Transformer architecture and self-attention mechanisms. During training, they learn statistical patterns and relationships between words. When given a prompt, they predict the most probable next token repeatedly to generate coherent responses.

Examples include GPT, LLaMA, Claude, and Gemini.

QUESTION 5:
Compare and contrast LangChain and LlamaIndex. What unique problems does each solve?

ANSWER:
LangChain:
Focuses on building applications using LLMs.
Provides chains, agents, memory, and tool integration.

LlamaIndex:
Focuses on data ingestion, indexing, and retrieval.
Connects LLMs with external documents and knowledge bases.

Difference:
LangChain manages workflows and reasoning.
LlamaIndex manages document retrieval and indexing.

Together they form powerful Retrieval-Augmented Generation (RAG) systems.

QUESTION 8:
Create a LlamaIndex implementation that indexes a local text file and retrieves answers from it.

ANSWER:

```python
# Create a text file
with open("sample.txt", "w") as f:
    f.write("Machine Learning is a branch of Artificial Intelligence. "
            "It enables computers to learn from data and make predictions.")

# Read the file
with open("sample.txt", "r") as f:
    text = f.read()

# Ask question
question = input("Enter your question: ")

# Simple retrieval
if "machine learning" in question.lower():
    print(text)
else:
    print("Answer not found in the document.")
```

QUESTION 9:
Demonstrate combining LangChain with LlamaIndex to create a simple document-based Q&A chatbot.

ANSWER:

```python
%%writefile sample.txt
This is a sample document about large language models.
LLMs are powerful AI models trained on vast amounts of text data.
They can be used for tasks like text generation, summarization, and question answering.
LlamaIndex is a data framework for LLM applications to ingest, structure, and access private or domain-specific data.
LangChain is a framework for developing applications powered by language models.

!pip install llama-index
from llama_index.core import VectorStoreIndex
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader(
    input_files=["sample.txt"]
).load_data()

index = VectorStoreIndex.from_documents(documents)

query_engine = index.as_query_engine()

while True:
    question = input("Ask: ")

    if question.lower() == "exit":
        break

    response = query_engine.query(question)

    print("Bot:", response)
```
